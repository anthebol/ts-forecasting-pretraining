```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Read the data
print("Loading data...")

# Replace commas with dots and parse correctly
df = pd.read_csv(
    'LD2011_2014.txt',
    sep=';',
    decimal=',',  # Specify that commas are used as decimal separators
    parse_dates=[0],  # Parse the first column as dates
    index_col=0,  # Set the first column as the index
    low_memory=False  # Handle large datasets efficiently
)

# Print initial data inspection
print("\nFirst few rows of the data:")
print(df.head())
print("\nData types of columns:")
print(df.dtypes)

# Basic plots without data transformation
plt.figure(figsize=(15, 10))

# 1. Plot first client's consumption over time
plt.subplot(2, 1, 1)
first_client = df.iloc[:, 0]  # Take first client
plt.plot(first_client.index, first_client.values, alpha=0.7)
plt.title('First Client Electricity Consumption Over Time')
plt.xlabel('Time')
plt.ylabel('Consumption (kW)')
plt.grid(True)

# 2. Plot first few clients
plt.subplot(2, 1, 2)
for i in range(min(5, len(df.columns))):
    plt.plot(df.index, df.iloc[:, i], label=f'Client {i}', alpha=0.7)
plt.title('First 5 Clients Consumption Patterns')
plt.xlabel('Time')
plt.ylabel('Consumption (kW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print basic info about the dataset
print("\nDataset Information:")
print(f"Number of clients: {len(df.columns)}")
print(f"Number of timestamps: {len(df)}")
print(f"Time range: {df.index.min()} to {df.index.max()}")
print("\nSample values from first client:")
print(df.iloc[:5, 0])
```
